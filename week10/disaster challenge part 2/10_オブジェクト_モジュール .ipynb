`for`や`while`によって繰り返し処理を行うことができます。例えば下のプログラムの例では、1～100までの和を計算します。

````python
#1～100までの和
sum=0
for ii in range(1,101):
  sum+=ii
print(sum)
````

では、たとえば、101～200までの和と201～300までの和も計算したい場合、どうしますか。

````python
#1～100までの和
sum=0
for ii in range(101,201):
  sum+=ii
print(sum)
sum=0
for ii in range(201,301):
  sum+=ii
print(sum)
````

このような形で、`for`を2個書いても良いですが、これは大変です。そこで自作の関数を作成します。

````python
def func_sum(a,b):
  sum = 0
  for ii in range(a,b+1):
    sum+=ii
  return sum
````

すると、次のように`for`を毎回書かなくても良くなります。

````python
print(func_sum(1,100))
print(func_sum(101,200))
````

こんな感じで、関数を自作すると、プログラムをスマートに書けるようになります。これが関数を自作することの御利益でした。

## オブジェクトとは？
　Pythonでは、あらゆる数値や文字列などの値は**オブジェクト**として扱われます。**オブジェクトはデータとそのデータを処理する関数をひとまとまりにしたもの**のことです。オブジェクトがもつ関数は「型」によって決まります。.formatはstr型オブジェクトがもつ関数、.appendはlist型オブジェクトがもつ関数です。これらのように、オブジェクトが持つ関数を**メソッド**と呼びます。
